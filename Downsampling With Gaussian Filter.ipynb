# Downsampling & Upsampling

## Downsampling

![](https://i.imgur.com/0zrrdKy.png)

```python
# read your image here and show it
# you should cover with the image and its shape
img = cv2.imread('image/bird.jpg')
b,g,r = cv2.split(img)
img_RGB = cv2.merge([r,g,b])
plt.imshow(img_RGB)
plt.show()
print('image shape: ', img_RGB.shape)
```

```python
# define a function that downsample the image with the input image matrix and the scale.
# the function will return a modefied image matrix.

def DownSamplingFunction(src,samplingScale):
  # plot the image and shape
  # -------------- To do ----------------- #
  move = int(1 / samplingScale)
  dst = src[0::move, 0::move]
  
  return dst.astype('uint8')
```

### Downsampling Without Gaussian Filter

```python
# plot your result with 2 scale, 1/4 and 1/8, also their shape.

# ----------- to do ------------------ #
D_withoutG_1 = DownSamplingFunction(img_RGB,1/4)
D_withoutG_2 = DownSamplingFunction(img_RGB,1/8)


# plot the image and shape
# -------------- To do ----------------- #
plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
plt.title("original")
plt.imshow(image)
plt.subplot(1,3,2)
plt.title("1/4")
plt.imshow(D_withoutG_1)
plt.subplot(1,3,3)
plt.title("1/8")
plt.imshow(D_withoutG_2)
plt.show()
print("original image:", image.shape)
print("1/4 image:", D_withoutG_1.shape)
print("1/8 image:", D_withoutG_2.shape)
```

## Upsampling

Chosing the 1/8 downsampling picture for upsampling
using Nearest-neighbor interpolation

```python
# ---------- to do 1, upsampling ------------ #
# using 1/8 image from previous work, using nearest neightbor method
upsampling = D_withG_2.repeat(8,axis=0).repeat(8,axis=1)
```

```python
# ---------- to do 2, plot ------------ #
# you should cover with the result image and the shape
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title("1/8")
plt.imshow(D_withG_2)
plt.subplot(1,2,2)
plt.title("Nearest Neighbor interpolation")
plt.imshow(upsampling)
plt.show()
print("1/8 shape:", D_withG_2.shape)
print("upsampling shape:", upsampling.shape)
```
